```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
from sklearn.cluster import KMeans

# Load necessary columns only
rating_df = pd.read_csv('/content/rating.csv', usecols=['movieId', 'rating'])
movie_df = pd.read_csv('/content/movie.csv', usecols=['movieId', 'genres'])

# Merge dataframes
merged_data = pd.merge(rating_df, movie_df, on='movieId', how='inner')

# Save the merged dataframe to disk
merged_data.to_csv('merged_data.csv', index=False)

# Read the merged dataframe from disk
merged_data = pd.read_csv('merged_data.csv')

# Function to perform KNN
def perform_knn(data, k):
    # Select features for KNN
    features = ['movieId', 'rating']

    # Prepare data for KNN
    X = data[features]

    # Fit KNN model
    knn = NearestNeighbors(n_neighbors=k)
    knn.fit(X)

    # Return KNN model
    return knn

# Function to visualize nearest neighbors and K-means clusters
def visualize_knn_and_kmeans(knn_model, kmeans_model, movie_id):
    # Find nearest neighbors of a specific movieId
    distances, indices = knn_model.kneighbors(merged_data[merged_data['movieId'] == movie_id][['movieId', 'rating']])

    # Get cluster labels
    cluster_labels = kmeans_model.labels_

    # Visualize both nearest neighbors and K-means clusters
    plt.figure(figsize=(12, 6))

    # Plot nearest neighbors
    plt.subplot(1, 2, 1)
    plt.plot(distances.squeeze(), marker='o', linestyle='-', color='b')
    plt.xlabel('Nearest Neighbor Index')
    plt.ylabel('Distance')
    plt.title('Nearest Neighbors of Movie ID {}'.format(movie_id))
    plt.grid(True)

    # Plot K-means clusters
    plt.subplot(1, 2, 2)
    plt.scatter(merged_data['movieId'], merged_data['rating'], c=cluster_labels, cmap='viridis', marker='.')
    plt.xlabel('Movie ID')
    plt.ylabel('Rating')
    plt.title('K-means Clusters')
    plt.colorbar(label='Cluster')

    plt.tight_layout()
    plt.show()

# Example usage
knn_model = perform_knn(merged_data, k=5)
kmeans_model = perform_kmeans(merged_data, num_clusters=3)
visualize_knn_and_kmeans(knn_model, kmeans_model, movie_id=1)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns

# Example user ratings data (replace with your actual data)
user_ratings = {
    1: {'movie1': 3, 'movie2': 4, 'movie3': 0, 'movie4': 5},
    2: {'movie1': 4, 'movie2': 3, 'movie3': 2, 'movie4': 0},
    3: {'movie1': 0, 'movie2': 5, 'movie3': 4, 'movie4': 3}
}

# Function to calculate Pearson correlation coefficient
def pearson_correlation(user1_ratings, user2_ratings):
    # Calculate mean ratings
    mean_user1 = np.mean(list(user1_ratings.values()))
    mean_user2 = np.mean(list(user2_ratings.values()))

    # Calculate covariance and standard deviations
    numerator = sum((user1_ratings[i] - mean_user1) * (user2_ratings[i] - mean_user2) for i in user1_ratings if i in user2_ratings)
    denominator_user1 = sum((user1_ratings[i] - mean_user1) ** 2 for i in user1_ratings)
    denominator_user2 = sum((user2_ratings[i] - mean_user2) ** 2 for i in user2_ratings)

    # Calculate Pearson correlation coefficient
    if denominator_user1 != 0 and denominator_user2 != 0:
        correlation = numerator / np.sqrt(denominator_user1 * denominator_user2)
    else:
        correlation = 0  # If denominator is zero, correlation is zero

    return correlation

# Function to perform collaborative filtering for multiple target users
def collaborative_filtering(user_ratings, target_users, k):
    # Initialize predicted ratings matrix
    predicted_ratings_matrix = {}

    # Calculate Pearson correlation coefficient between target users and all other users
    for target_user in target_users:
        correlations = {}
        for user, ratings in user_ratings.items():
            if user != target_user:
                correlation = pearson_correlation(user_ratings[target_user], ratings)
                correlations[user] = correlation

        # Sort correlations dictionary by correlation coefficient in descending order
        sorted_correlations = sorted(correlations.items(), key=lambda x: x[1], reverse=True)

        # Select top k similar users
        similar_users = sorted_correlations[:k]

        # Predict ratings for target user based on similar users' ratings
        predicted_ratings = {}
        for movie in user_ratings[target_user]:
            if user_ratings[target_user][movie] == 0:  # Predict rating only for unrated movies
                numerator = sum(user_ratings[user][movie] * correlation for user, correlation in similar_users)
                denominator = sum(correlation for _, correlation in similar_users if user_ratings[user][movie] != 0)
                if denominator != 0:
                    predicted_ratings[movie] = numerator / denominator
                else:
                    predicted_ratings[movie] = 0  # If denominator is zero, predicted rating is zero

        predicted_ratings_matrix[target_user] = predicted_ratings

    return predicted_ratings_matrix

# Example usage
target_users = [1, 2, 3]
k = 2
predicted_ratings_matrix = collaborative_filtering(user_ratings, target_users, k)

# Convert predicted ratings matrix to a dataframe for visualization
predicted_ratings_df = pd.DataFrame(predicted_ratings_matrix)

# Visualize the predicted ratings matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(predicted_ratings_df, annot=True, cmap='viridis')
plt.title('Predicted Ratings Matrix')
plt.xlabel('Movie')
plt.ylabel('User')
plt.show()
```

```python
# Function to generate hybrid recommendations based on weighted combination of individual methods
def hybrid_recommendations(knn_recommendations, kmeans_recommendations, collab_filtering_recommendations):
    # Combine recommendations from all methods
    all_recommendations = knn_recommendations + kmeans_recommendations + collab_filtering_recommendations

    # Count occurrences of each movie recommendation
    movie_counts = {}
    for movie in all_recommendations:
        if movie in movie_counts:
            movie_counts[movie] += 1
        else:
            movie_counts[movie] = 1

    # Sort movies by their occurrence counts in descending order
    sorted_movies = sorted(movie_counts.items(), key=lambda x: x[1], reverse=True)

    # Select top recommendations
    top_recommendations = [movie for movie, _ in sorted_movies[:2]]

    return top_recommendations, knn_recommendations, kmeans_recommendations, collab_filtering_recommendations

# Example usage
knn_recommendations = ['Movie1', 'Movie5', 'Movie6']
kmeans_recommendations = ['Movie4', 'Movie5', 'Movie6']
collab_filtering_recommendations = ['Movie5', 'Movie6', 'Movie1']

top_recommendations, knn_rec, kmeans_rec, collab_rec = hybrid_recommendations(knn_recommendations, kmeans_recommendations, collab_filtering_recommendations)
print("KNN Recommendations:", knn_rec)
print("K-means Recommendations:", kmeans_rec)
print("Collaborative Filtering Recommendations:", collab_rec)
print("Top Recommendations:", top_recommendations)
```

```python
import numpy as np

# Placeholder function for generating test data
def generate_test_data():
    return {1: np.random.choice(range(1, 11), size=5), 2: np.random.choice(range(1, 11), size=5)}

# Placeholder function for generating recommended movies
def generate_recommendations():
    return {1: np.random.choice(range(1, 11), size=3), 2: np.random.choice(range(1, 11), size=3)}

# Function to calculate RMSE
def calculate_rmse(predicted_ratings, actual_ratings):
    rmse = np.sqrt(np.mean([(predicted_ratings[user_id] - np.mean(actual_ratings[user_id])) ** 2 for user_id in actual_ratings.keys()]))
    return rmse

# Function to calculate precision and recall
def calculate_precision_recall(actual_data, recommended_data):
    precision_values = []
    recall_values = []
    for user_id, actual_movies in actual_data.items():
        recommended_movies = recommended_data[user_id]
        actual_set = set(actual_movies)
        recommended_set = set(recommended_movies)
        true_positives = len(actual_set.intersection(recommended_set))
        false_positives = len(recommended_set - actual_set)
        false_negatives = len(actual_set - recommended_set)
        precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) > 0 else 0
        recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) > 0 else 0
        precision_values.append(precision)
        recall_values.append(recall)
    return np.mean(precision_values), np.mean(recall_values)

# Example usage
test_data = generate_test_data()  # Generate test data
recommendations = generate_recommendations()  # Generate recommended movies
predicted_ratings = {user_id: np.random.choice(range(1, 11)) for user_id in test_data.keys()}  # Generate random predicted ratings

# Calculate RMSE
rmse_value = calculate_rmse(predicted_ratings, test_data)

# Calculate precision and recall
precision, recall = calculate_precision_recall(test_data, recommendations)

# Print results
print("RMSE:", rmse_value)
print("Precision:", precision)
print("Recall:", recall)
```
